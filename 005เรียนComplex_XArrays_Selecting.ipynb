อย่างไรก็ดี ยังทำไม่ได้ มีปัญหาตอนนำ data array มารวมเป็น data set กัน

```python
import numpy as np
import xarray as xr
import pandas as pd
```

```python
#กำหนดค่าละติจูดและลองจิจูดของข้อมูล
latitude = np.arange(20, 50.5, 0.5)
longitude = np.arange(-125, -65.5, 0.5)
print(np.shape(latitude), np.shape(longitude))
```

```python
#กำหนดระดับความกดอากาศ
levels = np.array([1000, 850, 500, 250])
print(np.shape(levels))
```

```python
#กำหนดความถี่รายชั่วโมง ด้วยการใช้ pandas
times = pd.date_range(start='2/1/21 00:00:00', end='2/1/21 23:00:00', freq='H')
print(np.shape(times))
```

```python
#กำหนดค่าข้อมูลแบบสุ่ม
temperature = np.random.random((24, 4, 61, 119)) #time hour=24, pressure level=4, latitude=61, longitude= 119
print(np.shape(temperature))
```

```python
#กำหนดระดับความสูงจากพื้นโลก 5 ระดับ
heights = np.array([0, 500, 1000,  1500, 2000])
print(np.shape(heights))
```

```python
wind_speed = np.random.random((24, 5, 61, 119))
print(np.shape(wind_speed))
```

```python
#สร้าง variable data array ของ tempertaure
temperature_da = xr.DataArray(temperature, coords=(times, levels, latitude, longitude),
                             dims=['time', 'level', 'latitude','longitude'])
#เพิ่ม meta data
temperature_da.name = 'Air Temperature'

#เพิ่ม attribute
temperature_da.attrs['units'] = 'degC'
```

```python
temperature_da
```

```python
##สร้าง variable data array ของ wind speed
wspd_da = xr.DataArray(wind_speed, coords=(times, heights, latitude, longitude),
                             dims=['time', 'height', 'latitude','longitude'])

#เพิ่ม meta data
wspd_da.name = 'Wind Speed'

#เพิ่ม attribute
wspd_da.attrs['units'] = 'knots'
```
