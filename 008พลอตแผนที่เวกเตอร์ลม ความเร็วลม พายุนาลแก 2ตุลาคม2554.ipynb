# พลอตแผนที่เวกเตอร์ลม ความเร็วลม พายุนาลแก 2ตุลาคม2554
* ลมเป็นข้อมูลเวกเตอร์ ประกอบด้วย ทิศทาง (dierection) และความเร็ว (wind speed) เป็นตัวแปรที่สำคัญในการขับเคลื่อนพลังงานจากที่แห่งหนึ่งไปยังพื้นที่แห่งหนึ่งบนโลก ทำให้เกิดสมดุลระหว่างพลังงานส่วนเกิน (surplus) และพลังงานส่วนขาด (deficit) ลมเป็นตัวนำพาพลังงานในรูปของความร้อนแฝงที่ถูกพาไปพร้อมกับไอน้ำ (water vapor) ในอากาศ เมื่อไอน้ำเหล่านั้นอยู่ในสภาวะที่เหมาะสมจะเกิดการกลั่นตัวเปลี่ยนสถานะไปเป็นหยาดน้ำฟ้า (precipitation) ไม่ว่าจะเป็นของเหลวหรือก๊าซ ก็จะมีการปลดปล่อยพลังงานออกมาจากการเปลี่ยนแปลงสถานะนี้ ให้กับสิ่งแวดล้อม ดังนั้นการศึกษาการเคลื่อนที่ของกระแสลม จึงเป็นหลักฐานชั้นดีทางวิทยาศาสตร์บรรยากาศที่จะช่วยให้เข้าใจพลวัตของพลังงานที่หมุนเวียนในช้ันบรรยากาศโลกได้
* ลมประกอบด้วย 2 คอมโพเนนท์ ที่เป็นตัวกำหนดทิศทางของลม นั่นคือ 
    * ลมในแนวตะวันออก-ตะวันตก หรือ u component ที่อยู่ในแนวแกน x  
        * ค่า u เป็น - แสดงว่าเป็นลมที่พัดมาจากทิศตะวันออก พัดไปทางทิศตะวันตก เรียกว่า easterly wind
        * ค่า u เป็น + แสดงว่าเป็นลมที่พัดมาจากทิศตะวันตก พัดไปทางทิศตะวันออก เรียกว่า westerly wind
    * ลมในแนวเหนือ-ใต้ หรือ v component ที่อยู่ในแนวแกน y 
        * ค่า v เป็น - แสดงว่าเป็นลมที่พัดมาจากทิศเหนือ พัดลงไปทางทิศใต้ เรียกว่า northerly wind
        * ค่า v เป็น + แสดงว่าเป็นลมที่พัดมาจากทิศใต้ พัดขึ้นไปทางทิศเหนือ เรียกว่า southerly wind

* อ่านเรื่องลม ทิศลม เวกเตอร์ http://colaweb.gmu.edu/dev/clim301/lectures/wind/wind-uv

* ปี 2554 เป็นปีที่ประเทศไทยได้รับผลกระทบจากมหาอุทกภัย เนื่องจากเป็นปีที่มีปริมาณน้ำที่มากกว่าปีปรกติ โดยที่มีจำนวนพายุโซนร้อนเคลื่อนตัวเข้าสู่พื้นที่อินโดจีนที่รวมทั้้งประเทศไทยเป็นจำนวนหลายลูกมากกว่าปีปรกติ นอกจากนั้นยังมีอิทธิพลของร่องมรสุมที่พาดผ่านปรากฏในหลายช่วงเวลา ดังนั้นปี 2554 จึงเป็นปีที่เหมาะแก่การนำมาศึกษาทางด้านบรรยากาศวิทยา ให้นิสิตอ่านบันทึกเหตุการณ์มหาอุทกภัยปี 2554 http://tiwrmdev.hii.or.th/current/menu.html
* ข้อมูล ERA5 daily จาก https://cds.climate.copernicus.eu/cdsapp#!/dataset/reanalysis-era5-pressure-levels?tab=overview
* อ่านเอกสารตัวแปรจาก https://confluence.ecmwf.int/display/CKB/ERA5%3A+data+documentation
* ERA5 มีความละเอียด 0.25 deg นั่นคือ ราวๆ 25 กิโลเมตรที่เส้นศูนย์สูตร
* ลมจาก ERA5 มีหน่วยเป็น m/s

## สิ่งที่นิสิตจะได้เรียนรู้
* อธิบายข้อมูล ตัวแปร ที่ได้จาก ERA5 แบบรายชั่วโมงแต่เลือกเฉพาะ 00 UTC สาธิตการใช้ข้อมูล โดยเลือกวันที่ 2-5 ตุลาคม 2011 เพื่อดูพายุนาลแก
* พลอตแผนที่เวกเตอร์ลมในแต่ละระดับ เพื่อวิเคราะห์ความผันแปรของทิศทางและความเร็วในแต่ละระดับความกดอากาศ
* สกัดข้อมูลความเร็วลม u component เพื่อนำไปวิเคราะห์ถึงความผันแปรของทิศลมในแต่ละระดับความกดอากาศ หรือที่เรียกว่า wind shear ตามพิกัดภูมิศาสตร์และแต่ละระดับความกดอากาศที่ต้องการ เพื่อนำออกไปวิเคราะห์ต่อในโปรแกรมเช่น excel หรือ R
* wind shear หรือ ลมเฉือน ประกอบไปด้วยการเปลี่ยนแปลงในสองตัวแปร คือ 
    * ทิศทางลมในแต่ละระดับความกดอากาศ เรียกว่า directional wind shear
    * ความเร็วในแต่ละระดับความกดอากาศ  เรียกว่า speed wind shear 
* อ่านเรื่อง wind shear

    *ในทางการบิน https://www.youtube.com/watch?v=jXxr4VHqlSw และเลคเชอร์ windshear สำหรับนักบิน https://www.youtube.com/watch?v=FxXwqAm1a-Y

* โค้ดในการคำนวณหา wind shear จากโมเดล โดย ucar https://unidata.github.io/python-gallery/examples/Wind_Shear_Vectors_Example.html
  

## 1.จัดการข้อมูล ERA5 ที่ได้ดาวน์โหลดมา

```python
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
from cartopy import feature as cf
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#อ่านข้อมูล
df = xr.open_dataset('era5_daily_2_5oct_2011_globe.nc')
print(df)
```

### เลือกข้อมูลให้เล็กลงเน้นเฉพาะพื้นที่อินโดจีนและบริเวณโดยรอบ และทำการประมาณค่าใหม่ให้เหลือ 0.5 deg เพื่อทำการประมวลผลได้เร็วขึ้น 

```python
df = df.sel(longitude=np.arange(60, 130, 0.5), latitude=np.arange(-5, 45, 0.5), method='nearest') # 0.5 คือกริดใหม่ที่ต้องการจากการประมาณค่าด้วย nearest
```

```python
df
```

### สังเกตุค่า coordinates ลดลง และระยะห่างระหว่างกริดเปลี่ยนไปจาก 0.25 เป็น 0.5 deg 
* ต่อไปสิ่งที่จะใช้ค่าตัวแปร u และ v โดยที่เป็นตัวแปรของทิศที่ลมพัดมา ในแนวแกน x และ y ตามลำดับ 
    * หากลม u มีค่าเป็น + แสดงว่าลมพัดมาจากทิศตะวันตก หรือ westerly wind  แต่หาก u มีค่าเป็น - แสดงว่าลมพัดมาจากทิศตะวันออก หรือ easterly wind
    * หากลม v มีค่าเป็น + แสดงว่าลมพัดมาจากทิศใต้ หรือ southerly wind  แต่หาก v มีค่าเป็น - แสดงว่าลมพัดมาจากทิศเหนือ หรือ northerly wind


* อ่านต่อจาก http://colaweb.gmu.edu/dev/clim301/lectures/wind/wind-uv

```python
df=df.sel(time='2011-10-02', level=850)
```

```python
df
```

```python
x = df.longitude
y = df.latitude
u = df.u
v =df.v
```

## 2.ฝึกพลอตแผนที่ทิศทางลม

```python
#ฝึกการพลอตแผนที่ด้วย cartopy+matplotlib
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
```

```python
#ฝึกการพลอตแผนที่ด้วย cartopy+matplotlib ด้วยการเพิ่มขอบเขตประเทศและกำหนดขอบเขตแผนที่
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cf.BORDERS) #เพิ่มขอบเขตประเทศ
ax.set_extent([75, 125, 0, 25]) # กำหนดขอบเขตแผนที่
```

```python
# พลอตลูกศรลม แต่เหมือนว่ามันหนาแน่นเกินไป เพราะความละเอียดข้อมูล ERA5 อยู่ที่ 0.25 deg  แต่ในที่นี้เราลดเหลือ 0.5 deg แล้ว
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cf.BORDERS)
ax.set_extent([75, 125, 0, 25])
ax.barbs(x, y, u, v)
plt.show()
```

```python
#ลดการแสดงผลลูกศรลม ให้พลอตทุกๆ 5 ช่องกริด
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cf.BORDERS)
ax.set_extent([75, 125, 0, 25])
ax.barbs(x[::3], y[::3], u[::3,::3], v[::3,::3]) #ให้พลอตทุกๆ 3ช่องกริด
```

```python
# พลอตเวกเตอร์ลม
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cf.BORDERS)
ax.set_extent([75, 125, 0, 25])
ax.quiver(x[::3], y[::3], u[::3,::3], v[::3,::3], scale=500)

```

## 3. พลอตแผนที่เวกเตอร์ลมและความเร็วลม

```python
#คำนวณความเร็วลม wind speed
ws = np.sqrt(u**2+v**2)
```

```python
plt.figure(figsize=(13,6.2))

ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cf.BORDERS)
ax.set_extent([75, 125, 0, 25])
ax.set_title('wind vector of 2 October 2011 at 850 hPa' )


max_ = np.round((np.max(ws)))+2
min_ = 0
#min_ = np.round((np.min(ws))) 
vstep = 1.0
levels = np.arange(min_, max_, vstep)
cwf = ax.contourf(x, y, ws, levels, cmap='jet',
                 extend='both', transform=ccrs.PlateCarree())

#พลอต color bar
# คำนวณมิติข้อมูล (height_of_image / width_of_image)
im_ratio = ws.shape[0]/ws.shape[1]
cbar = plt.colorbar(cwf,fraction=0.047*im_ratio)
cbar.set_label('[m/s]', fontsize=15)

#wind vector
Q = ax.quiver(x[::3], y[::3], u[::3,::3], v[::3,::3],  scale=30,  color='red', pivot='mid', units='inches')

#scale wind vector 
#qk = ax.quiverkey(Q, 0.9, 0.9, 20, r'$1 \frac{m}{s}$', labelpos='E', coordinates='figure')

#เซตค่าชื่อกริด การแสดงกริดเส้นโครง
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
```

#### งานมอบหมายนิสิต 
* ให้นิสิตทำการพลอตดูแผนที่ลมในแต่ละระดับความกดอากาศ เพื่อวิเคราะห์เปรียบเทียบกัน 1000hPa, 925hPa, 850hPa, 500hPa, 200hPa แล้วสังเกตุถึงค่าความเร็วและทิศทางของลมในพื้นที่ที่สนใจ เช่นในแต่ละประเทศ ในพื้นที่ทะเล มหาสมุทร แล้วทำการบรรยายออกมาเป็นตัวหนังสือ

## 4.สกัดค่าความเร็วลม U หรือลมที่พัดในแนวตะวันออก-ตะวันตก ในแต่ละระดับความกดอากาศจากค่าพิกัดที่สนใจ


```python
u
```

```python
#อ่านข้อมูลเข้ามาใหม่
dfu = xr.open_dataset('era5_daily_2_5oct_2011_globe.nc')

# ประมาณค่ากริดใหม่จาก 0.25 เป็น 0.5 คือกริดใหม่ที่ต้องการจากการประมาณค่าด้วย nearest
da = dfu.sel(longitude=np.arange(60, 130, 0.5), latitude=np.arange(-5, 45, 0.5), method='nearest') 
```

```python
da
```

```python
#เลือกลมทุกระดับของวันที่ 2
da=da.sel(time='2011-10-02')

# เก็บตัวแปร u
uwind = da.u
```

```python
uwind
```

### สกัดค่าลม u พิกัดจากค่าพิกัดภูมิศาสตร์

```python
# กำหนดค่าเส้นโครงแผนที่ของข้อมูล 
datacrs = ccrs.PlateCarree()

# แปลงค่าจุดที่ต้องการสกัดค่าตัวแปรในแต่ละระดับเป็นแบบเส้นโครงแผนที่ Plate Carree for lat/lon grid เสมอ
x, y = datacrs.transform_point(100, 15.0, src_crs=ccrs.PlateCarree())

# สกัดค่าอุณหภูมิในแต่ละระดับความกดอากาศ
u_ext=uwind.sel(longitude=x, latitude=y) 
```

```python
u_ext
```

```python
# พลอตข้อมูลอย่างรวดเร็ว
u_ext.plot()
```

### สังเกตุว่า 
* โดยปรกติแล้วในช่วงมรสุมหรือในช่วงฤดูฝนของประเทศไทย จะมีลมมรสุมตะวันตกเฉียงใต้เป็นลมหลักในลักษณะลมใกล้ผิวพื้นจนถึงระดับชั้นกลางของบรรยากาศชั้นโทโพสเฟียร์ ในขณะที่ลมชั้นบนจะมีทิศทางตรงกันข้าม นั่นคือเป็นลมตะวันออก

* ลม u ในระดับตั้งแต่ 1000-500 mb จะมีค่าเป็น + นั่นคือ ลมในวันที่ 2 ตุลาคม 2011 องค์ประกอบลม u เป็นลมที่พัดมาจากทิศตะวันตก หรือ westerly wind component ในระดับล่างถึงกลางของชั้นบรรยากาศ  เนื่องจากมีศูนย์กลางของพายุหมุนเขตร้อนอยู่ที่อ่าวตังเกี๋ย ลมด้านล่างจึงพัดเข้าหาศูนย์กลางพายุ หรือที่เรียกว่า convergecne 

* ในขณะที่ลม u ตั้งแต่ระดับ 400-200 mb มีค่าเป็น - แสดงว่าเป็นลมที่พัดออก หรือ divergence ที่มีทิศพัดมาจากทิศตะวันออกหรือ easterly wind
