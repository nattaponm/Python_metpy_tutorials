# ดึงข้อมูล upper air ด้วย Siphon แพคเกจ
* ข้อมูลการหยั่งอากาศในระดับบนที่ใช้เครื่องมือ เรดิโอซอล (radiosonde) ที่ติดอุปกรณ์เซนเซอร์ตรวจวัดตัวแปรในแต่ละระดับชั้นบรรยากาศ ทำให้ทราบถึงลักษณะกายภาพของบรรยากาศในแต่ละระดับความกดอากาศ ช่วยเอื้อต่อการนำมาวิเคราะห์และติดตามสภาพอากาศที่รุนแรง เพื่อพยากรณ์อากาศล่วงหน้าได้ เป็นข้อมูลที่นักอุตุนิยมวิทยาต้องการเป็นอย่างมาก 

* rawindsonde คือข้อมูลตำแหน่งของเรดิโอซอล ที่ถูกระบุตำแหน่งขณะที่ลูกบอลลูนที่นำพาอุปกรณ์เซนเซอร์ขึ้นไปนั้นในแต่ละระดับความสูง ทำให้ได้ข้อมูลความเร็วและทิศทางลม ("radar wind -sonde") https://en.wikipedia.org/wiki/Radiosonde

* siphon package https://unidata.github.io/siphon/latest/examples/upperair/Wyoming_Request.html#
* แหล่งรวมข้อมูล upperair ของ wyoming 
    * ดาวน์โหลดข้อมูล upper air จาก https://weather.uwyo.edu/upperair/sounding.html
    * สร้างแผนที่ upper air ทั่วโลกได้ https://weather.uwyo.edu/upperair/uamap.shtml

```python
# ดึงข้อมูลด้วย siphon แพคเกจ
from datetime import datetime #datetime object
```

```python
date = datetime(2017, 6, 22, 0) #Cindy https://en.wikipedia.org/wiki/Tropical_Storm_Cindy_(2017)
station = 'LCH' #Luoisiana จากเวป https://weather.uwyo.edu/upperair/sounding.html  
```

```python
from siphon.simplewebservice.wyoming import WyomingUpperAir
```

```python
df = WyomingUpperAir.request_data(date, station) # ดึงข้อมูลตามวันเวลาและสถานีเข้ามาเก็บใน dataframe ของ pandas
```

```python
df
```

```python
# ดูหน่วย
df.units
```

ดิกชันนารีที่เก็บตัวแปรและหน่วยของตัวแปร

```python
# หาว่าหน่วยของ dewpoint คืออะไร
df.units['dewpoint']
```

```python
#ต่อไปจะใช้ method หน่วยใน metpy --- แพคเกจ metpy จะเน้นเรื่องหน่วยมาก เพราะเป็นงานทางด้านวิทยาศาสตร์ที่สำคัญ การใช้หน่วยผิดพลาด ส่งผลร้ายแรงต่อการวิเคราะห์
from metpy.units import units
```

```python
# ทดลองเก็บค่าความกดอากาศจากคอลัมน์ pressure โดยให้นำหน่วยที่อยู่ในดิกชันนารีของข้อมูลมาด้วย
p = df['pressure'].values * units(df.units['pressure'])
```

```python
p
```

```python
# เก็บค่าอื่นๆ ที่ต้องใช้ในการพลอตค่า sounding ใน skew-T/log P ไดอะแกรม
p = df['pressure'].values * units(df.units['pressure'])
T = df['temperature'].values * units(df.units['temperature'])
Td = df['dewpoint'].values * units(df.units['dewpoint'])
u = df['u_wind'].values * units(df.units['u_wind'])
v = df['v_wind'].values * units(df.units['v_wind'])
```
